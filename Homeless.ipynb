```python
#Import dependencies/libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
#Read and Merge MD_Homeless_Per_County and MD_County_homeless csvs
df1 = pd.read_csv("Resources/MD_Homeless_PerCounty.csv")
df2 = pd.read_csv("Resources/MD_County_homeless.csv")
df3 = pd.merge(df1, df2, how="left", on=["FIPS_Code","FIPS_Code"])
df3.head()
```

```python
#Read and MergeMaryland_Counties_Socioeconomic_Characteristicsy and Resources/2015-2020-HIC-Counts csvs
df4 = pd.read_csv("Resources/Maryland_Counties_Socioeconomic_Characteristics.csv")
df5 = pd.read_csv("Resources/2015-2020-HIC-Counts.csv")
df6 = pd.merge(df4, df5, how="left", on=["FIPS_Code","FIPS_Code"])
df6.head()
```

```python
#Merge two dataframes
df =  pd.merge(df3, df6, how="left", on=["FIPS_Code","FIPS_Code"])

df.head()
```

```python
homeless_col=['Civilian_labor_force', 'Employed',
       'Unemployed', 'Unemployment_rate', 'Median_Household_Income',
       'Med_HH_Income_Percent_of_State_Total', 'Overall Homeless 2019',
       'Overall Homeless 2018', 'Overall Homeless 2017',
       'Overall Homeless 2016', 'Overall Homeless 2015',
       'Overall Homeless - Under 18 ', 'Overall Homeless - Age 18 to 24',
       'Overall Homeless - Over 24', 'Overall Homeless - Female',
       'Overall Homeless - Male', 'Overall Homeless - Non-Hispanic',
       'Overall Homeless - Hispanic/Latino', 'Overall Homeless - White',
       'Overall Homeless - Black or African American',
       'Overall Homeless - Asian',
       'Overall Homeless - American Indian or Alaska Native',
       'Overall Homeless - Native Hawaiian or Other Pacific Islander',
       'Overall Homeless - Multiple Races']
```

```python
for i in homeless_col:
    try:
        #print(i)
        #print(homeless_data_complete[i])
        df[i]= df[i].str.replace(',', '').astype(float)
    
    except:
        #print("Error")
        pass
```

```python
df.head()
```

```python
by_race_df = df[["County_name","Overall Homeless - White","Overall Homeless - Black or African American","Overall Homeless - Asian","Overall Homeless - American Indian or Alaska Native","Overall Homeless - Native Hawaiian or Other Pacific Islander","Overall Homeless - Multiple Races"]]
```

```python
# plot data in stack manner of bar type
by_race_df.plot(x='County_name', kind='bar', stacked=True,
        title='Race of Homeless people by MD Counties in 2019').legend(loc='upper center', title="Races")
plt.xlabel("County Name")
plt.ylabel("# of Homeless")
#plt.savefig("Resources/Race of Homeless people by MD Counties in 2019.png")
```

```python
#BlackN = int(homeless_data_complete["Overall Homeless - Black or African American"])

White = df["Overall Homeless - White"].sum()
Black = df["Overall Homeless - Black or African American"].sum()
Asian = df["Overall Homeless - Asian"].sum()
AmerInd = df["Overall Homeless - American Indian or Alaska Native"].sum()
NatHaw = df["Overall Homeless - Native Hawaiian or Other Pacific Islander"].sum()
MultiRace = df["Overall Homeless - Multiple Races"].sum()
```

```python
homelessMD = [White, Asian, Black, NatHaw, MultiRace, AmerInd]

labels = "White", "Asian", "Black or African American",  "Native Hawaiian or Other Pacific Islander",   "Multiple Race", "American Indian or Alaska Native",
colors = ['lightskyblue', "red", "blue", "lightgreen", "pink", "orange"]
subplots='true',
plt.figure(figsize = (10,8))
explode = (0.05, 0.05, 0.05, 0.1, 0.05, 0.1) 

fig1, ax1 = plt.subplots()

ax1.pie(homelessMD, labels=labels,  explode=explode, colors=colors, autopct='%5.1f%%', startangle=40, radius= 0.9, shadow=False)

ax1.axis('equal')
plt.title("Homelessness by Race in MD counties 2019")
#plt.savefig("Resources/Homeless by MD counties - pie.png")
plt.show()

```
